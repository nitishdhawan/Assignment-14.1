```python
# Problem 2: Write two basic update queries
```

```python
tbl2 = table("tblAdult", column("FnlWgt"), column("Sex"), column("Race"),column("Education"),column("Age"))
```

```python
rs = con.execute("SELECT * FROM tblAdult where FnlWgt=77516")
df = pd.DataFrame(rs.fetchmany(size=10))
df.columns = rs.keys()
df
```

```python
# First Update Query
stmt = tbl2.update().\
            where(tbl2.c.FnlWgt==77516).\
            values(Race='New Race')
```

```python
con.execute(stmt)
```

```python
rs = con.execute("SELECT * FROM tblAdult where FnlWgt=77516")
df = pd.DataFrame(rs.fetchmany(size=10))
df.columns = rs.keys()
df
```
